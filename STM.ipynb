```python
import pandas as pd
import os
import tqdm
```

```python
origin = pd.read_excel("C:/Users/zabor/Desktop/preprocess/depression_data_2020covid_deprred_clean_v5.xlsx")
```

```python
in_folder = 'C:/Users/zabor/Desktop/LDA/Data/txts_contwrd'
in_path = os.path.join(in_folder)
in_files = [f for f in os.listdir(in_path) if os.path.isfile(os.path.join(in_path, f))]

#összes fájl listába rakása
txt = []
for filename in tqdm.tqdm(in_files):
    try:
        with open(os.path.join(in_path, filename), 'r', encoding='utf-8') as infile:
            article = infile.read()
        txt.append(article)
    except Exception as e:
        print(e)
```

```python
origin["txts_contwrd"] = txt
```

```python
origin.to_csv("pre_stm.csv", index = False)
```

```python
stm = pd.read_csv("pre_stm.csv", sep = ",", encoding = "utf8")
```

```python
contwrd_hossz = []
for i in range(len(stm["txts_contwrd"])):
    contwrd_hossz.append(len(str(stm["txts_contwrd"][i]).split()))
stm["contwrd_hossz"] = contwrd_hossz
```

```python
final_stm = stm[stm["contwrd_hossz"] >= 5]
final_stm = final_stm.reset_index(drop=True)
```

```python
final_stm.to_csv("final_stm.csv", index = False)
```
